```python
# Usage of Ensure annotations
def get_product(x:int, y:int) -> int:
    return x*y
```

```python
get_product(x=3,y=3)
```

```python
# undesirable wrong output due to flawed logic
get_product(x=3, y="3")
```

```python
import unittest

# Missing attribute patched, so 'ensure' doesn't crash
if not hasattr(unittest.TestCase, 'assertRaisesRegexp'):
    unittest.TestCase.assertRaisesRegexp = unittest.TestCase.assertRaisesRegex

from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_prodict(x:int, y:int) -> int:
    return x*y
```

```python
get_prodict(x=2, y=3)
```

```python
get_prodict(x=2, y="3")
```
